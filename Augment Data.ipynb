```
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from keras.optimizers import Adam, SGD, RMSprop 
from keras.losses import binary_crossentropy
from keras import regularizers
from keras.callbacks import EarlyStopping, ModelCheckpoint, CSVLogger
from keras import backend as K
```

```
from matplotlib import pyplot as plt
import numpy as np
from os import listdir
```

```
root = "/content/drive/My Drive/Colab Notebooks/dataset/"
```

```
datagen = ImageDataGenerator(rotation_range=30, 
                             width_shift_range=0.2, 
                             height_shift_range=0.2, 
                             shear_range=0.1, 
                             zoom_range=0.1, 
                             fill_mode='nearest',
                             horizontal_flip=True, 
                             vertical_flip=True)
```

```
hotdog = listdir(root+"train data/1")
not_hotdog = listdir(root+"train data/0")
```

```
for i in range(len(not_hotdog)):
  img = img_to_array(load_img(root+"train data/0/"+not_hotdog[i]))
  img = img.reshape((1,)+img.shape)
  count = 0
  img_flow = datagen.flow(img, batch_size=1)
  for j, new_img in enumerate(img_flow):
    new_img = array_to_img(new_img[0], scale=True)
    new_img.save(root+"aug data 2/0/"+str(i)+"_"+str(count)+".jpeg")
    count += 1
    print ("For image " + str(i) + ", " + str(count) + "th image is created")
    if (count == 30):
      break;
```
